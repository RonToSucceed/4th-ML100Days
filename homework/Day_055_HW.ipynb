- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(5)

%matplotlib inline
```

```python
estimators = [('k_means_8', KMeans(n_clusters=8)),
              ('k_means_3', KMeans(n_clusters=3)),
              ('k_means_bad_init', KMeans(n_clusters=3, n_init=1,
                                               init='random'))]
```

```python
fignum = 1
titles = ['8 clusters', '3 clusters', '3 clusters random init']
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(4,3))
    axis = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
    #fit data
    est.fit(X)
    labels = est.labels_
    
    axis.scatter(X[:,3], X[:,0],X[:,2],
                 c=labels.astype(np.float), edgecolor='k')
    
    axis.w_xaxis.set_ticklabels([])
    axis.w_yaxis.set_ticklabels([])
    axis.w_zaxis.set_ticklabels([])
    axis.set_title(name) #titles[fignum-1]
    axis.dist = 12
    fignum = fignum+1
#Plot the ground truth
fig = plt.figure(fignum, figsize=(4,3))
axis = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

for name, label in [('cls0', 0),('cls1', 1),('cls2', 2)]:
     axis.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean(),
              X[y == label, 2].mean() + 2, name,
              horizontalalignment='center',
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))
        
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)
axis.scatter(X[:, 3], X[:, 0], X[:, 2], c=y, edgecolor='k')

axis.w_xaxis.set_ticklabels([])
axis.w_yaxis.set_ticklabels([])
axis.w_zaxis.set_ticklabels([])
axis.set_title('Ground Truth')
axis.dist = 12
```
